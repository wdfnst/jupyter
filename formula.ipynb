```python
import numpy as np
from math import exp, expm1, sin, cos, pi
from matplotlib import pyplot as plt

x = np.arange(1, 100)
sin_y = []
cos_y = []
for i in range(51):
    '''sin_y.append([sin(2 * pi * j)/ 51 for j in x])
    cos_y.append([cos(2 * pi * j)/ 51 for j in x])'''
    sin_y.append(np.sin(2*np.pi*x))
    cos_y.append(np.cos(2*np.pi*x))

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(221)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(222)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.subplot(223)
plt.plot(x, sin_y[0], 'r--')

plt.subplot(224)
plt.plot(x, cos_y[30], 'r--')



fig, axs = plt.subplots(2,5, figsize=(15, 6), facecolor='w', edgecolor='k')
axs = axs.ravel()    
y_cos = []
for i in range(10):
    axs[i].plot(t2, np.cos(2*np.pi*t2/(i + 1)), 'r--')
    y_cos.append(np.cos(2*np.pi*t2/(i + 1)))

yy = np.sum(y_cos, axis=0)
print(yy)

fig, axs = plt.subplots(1,1, figsize=(15, 6), facecolor='w', edgecolor='k')
plt.plot(t2, yy, 'r--')
plt.show()
```
